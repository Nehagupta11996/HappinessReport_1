```python
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.model_selection import train_test_split
```

imported the libraries required for machine learning of the data

As this is linear regression problem imported the linear regression libraries 

```python
df_new.keys()
```

```python
df_new=df_new.reindex(columns=["Region","Standard Error","Economy (GDP per Capita)","Family","Health (Life Expectancy)","Freedom","Trust (Government Corruption)","Generosity","Dystopia Residual","Happiness Score"])
```

```python
df_new.keys()
```

```python
x=df_new.iloc[:,0:-1]
x.head()
```

```python
y=df_new.iloc[:,-1]
y.head()
```

separated the x and y data

```python
x.shape
```

```python
y.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
lm=LinearRegression()
```

Making the instance of the class

```python
lm.fit(x_train,y_train)
```

training data 

```python
lm.coef_
```

```python
lm.intercept_
```

```python
df_new.columns
```

```python
lm.score(x_train,y_train)
```

```python
#predict the value
pred=lm.predict(x_test)
print("predicted result price:",pred)
print("actual price",y_test)
```

```python
print("error:")
print("mean absolute error:", mean_absolute_error(y_test,pred))
print("mean squared error:", mean_squared_error(y_test,pred))

print("root Mean Squared error:",np.sqrt(mean_squared_error(y_test,pred)))
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,pred))
```
